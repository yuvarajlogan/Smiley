```python
from turtle import*
import turtle

pen= turtle.Turtle()
turtle.clear()

pen.width(10)
pen.fillcolor('black')
pen.begin_fill()
pen.circle(100)
pen.up()


#def of EYE
def eye(col,rad):
    pen.down()
    pen.fillcolor(col)
    pen.begin_fill()
    pen.width(4)
    pen.circle(rad)
    pen.up()
    
 #fill color

pen.fillcolor('pink')
pen.begin_fill()
pen.circle(100)
pen.end_fill()
pen.up()

#outer face
#pen.width(10)
#pen.fillcolor('black')
#pen.begin_fill()
#pen.circle(100)
#pen.up()

#fn of eye
pen.goto(-40, 120)
eye('red', 15)
pen.goto(-37,125)
eye('red',5)
pen.goto(40, 120)
eye('red', 15)
pen.goto(40,125)
eye('white',5)

pen.goto(-40, 85)
pen.down()
pen.right(90)
pen.circle(40, 180)
pen.up()

# draw tongue
pen.goto(-10, 45)
pen.down()
pen.right(180)
pen.fillcolor('red')
pen.begin_fill()
pen.circle(10, 180)
pen.end_fill()
pen.hideturtle()

pen.clear()
```
